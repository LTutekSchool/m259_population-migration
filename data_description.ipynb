# Key Takeaways for Dataset (USA)
* Population Growth: The population of the United States has a wide range and substantial variability, with values ranging from 180 million in 1960 to 334.9 million in 2023. This suggests a long-term upward trend, but significant fluctuations in the population growth over time. If population is used as a feature in modeling, normalization or scaling might be necessary to account for the large range of values.
* Net Migration: The netMigration column has both positive and negative values, reflecting periods of immigration (positive) and emigration (negative). The migration trend has steadily increased, with net migration reaching over 1.8 million in 2023. The dataset suggests a complex picture of how migration influences the U.S. population, with net migration generally rising despite the overall population increase.
* Population in Millions: The population_in_millions column shows a skewed distribution, as seen in the quartile values (the 25th percentile is at 217 million, and the 75th percentile is at 302 million). The maximum value of 334 million indicates a strong upward trend, yet a large portion of the data is skewed towards lower population values. This could be important for understanding the dynamics of population distribution.

```python
import matplotlib.pyplot as plt

# Filter the dataset for rows where the country is "United States"
usa_data = df[df['country'] == 'United States']

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot population on the primary y-axis
ax1.plot(usa_data['year'], usa_data['population'], marker='o', linestyle='-', color='red', label='Population')
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Population', fontsize=14, color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for net migration
ax2 = ax1.twinx()
ax2.plot(usa_data['year'], usa_data['netMigration'], marker='o', linestyle='-', color='blue', label='Net Migration')
ax2.set_ylabel('Net Migration', fontsize=14, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and layout
plt.title('Impact of Immigration (Net Migration) on Population Growth in the United States Over Time', fontsize=16)
fig.tight_layout()

# Show the plot
plt.show()


```
